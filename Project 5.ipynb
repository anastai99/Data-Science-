## Project 5 Discord Logs

```python
import json
import pandas as pd
from datetime import datetime

#opening the json file containing discord logs
f = open('discord_logs.json',)
  
#loading json file
data = json.load(f)

#creating dictionary for new dataframe containing discord logs
directory = []
#looping thru all the events happening in discord
for i in range(len(data['messages'])):
    listing = {}
    #extracting when someone joined or left voice channel from discord
    if ('joined voice channel' in data['messages'][i]['embeds'][0]['description'] or 'left voice channel ' in data['messages'][i]['embeds'][0]['description']):
        listing['Name'] = data['messages'][i]['embeds'][0]['author']['name'] 
        listing['Event'] = data['messages'][i]['embeds'][0]['description'][11:][:4]
        listing['Time'] = data['messages'][i]['embeds'][0]['timestamp']
        #adding it too dictionary

        directory.append(listing)
#made a dataframe from the dictionary after the for loop has went thru all the events happening in discord
df = pd.DataFrame(directory)
df
```

```python
#retrieved all of the unique names
names = df["Name"].unique()
#new dictionary to hold information of how long one stayed in discord
directory2 = []
#looping thru each name and analyzing each one of them
for i in names:
    listing = {}
    #'a' is a dataframe showing all the time this person joined the server
    a = df[(df['Name'] == i) & (df['Event'] == 'join')].reset_index()
    #'b' is a dataframe showing all the time this person left the server
    b  = df[(df['Name'] == i) & (df['Event'] == 'left')].reset_index()
    #creating new column in datafram 'a' which says when the person left called 'Time_left'
    a["Time_left"] = b["Time"]

    #converting the time and time_left columns to time format to make arthemic easier 
    a["Time"] = pd.to_datetime(a["Time"]).dt.tz_convert("America/New_York")
    a["Time_left"] = pd.to_datetime(b["Time"]).dt.tz_convert("America/New_York")
    
    #necolumn in a called duration which shows time stayed in a day
    a["duration"] =  a["Time_left"] -a["Time"]
    
    #adding name and sum of all the times stayed in server to dictionary
    listing['Name'] = i
    listing["time_in_channel"] = a['duration'].sum()
    
    
    directory2.append(listing)
#created new dataframe 
df2 = pd.DataFrame(directory2)
display(df2)
f.close
```
