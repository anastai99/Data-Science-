## Top 50 Male runners

I've used the request library to get the HTML version of a website. The website contained links that had information about each of the runners. Using the `BeautifulSoup` library I was able to parse information for each runner. This was done by a for loop retrieving the links for each runner and requesting that web page with the request library. The request library gave the HTML containing their specific information about themselves. Every time the for loop ran it added the needed information to a dictionary. After the for loop was finished, I created a Data Frame for the top 50 males with the Pandas Library. There was also a 2 second delay between each request to each runnerâ€™s site.

```python
#importing request, BeaitifulSoup, sleep and Pandas
import requests
from bs4 import BeautifulSoup
from time import sleep
import pandas as pd
```

```python
#requesting url containing information
r = requests.get("https://results.chicagomarathon.com/well-known/2019/?page=1&content=well-known&event=MAR&lang=EN_CAP&num_results=50&pid=list&search%5Bsex%5D=M&search%5Bage_class%5D=%25").text
```

```python
#using BeaitfulSoup to get parse HTML file
soup = BeautifulSoup(r)
names= soup.find_all('h4')
#dictoray that'll contain runners infromation
directory = []
#forloop runs 50 times so it gets 50 runners
for i in range(0,50):
    listing = {}
    #finding link of the runner then requesting it with url
    a = names[i].find("a")['href']
    name_site = "https://results.chicagomarathon.com/well-known/2019/"+a
    r = requests.get(name_site).text
    #2 second delay
    for i in range(2):
        sleep(i)
    
    #gets all information needed and adding to dictionary
    soup = BeautifulSoup(r)
    info= soup.find_all('div', class_="detail-box box-general")
    splits= soup.find_all('div', class_="table-responsive")
    
    listing["Name CTZ"] = info[0].find('td',class_="f-__fullname last").get_text()
    listing["Age Group"] = info[0].find('td',class_="f-age_class last").get_text()
    listing["Bib Number"] = info[0].find('td',class_="f-start_no_text last").get_text()
    listing["Age"] = info[0].find('td',class_="f-age last").get_text()
    listing["City, State"] = info[0].find('td',class_="f-__city_state last").get_text()
    
    
    Fivek = splits[0].find_all('tr',class_="f-time_05 split")
    Tenk = splits[0].find_all('tr',class_="list-highlight f-time_10 split")
    Fifteenk = splits[0].find_all('tr',class_="f-time_15 split")
    twentyk = splits[0].find_all('tr',class_="list-highlight f-time_20 split")
    Half = splits[0].find_all('tr',class_="f-time_52 split")
    twentyfive = splits[0].find_all('tr',class_="list-highlight f-time_25 split")
    thirty = splits[0].find_all('tr',class_="f-time_30 split")
    thirtyfive = splits[0].find_all('tr',class_="list-highlight f-time_35 split")
    fourty = splits[0].find_all('tr',class_="f-time_40 split")
    finish = splits[0].find_all('tr',class_="list-highlight f-time_finish_netto highlight")
    
    
    
    
    listing["05K"] = Fivek[0].find('td',class_='time').get_text()
    listing["10K"] = Tenk[0].find('td',class_='time').get_text()
    listing['15K'] = Fifteenk[0].find('td',class_='time').get_text()
    listing["20K"] = twentyk[0].find('td',class_='time').get_text()
    listing["Half"] = Half[0].find('td',class_='time').get_text()
    listing['25K'] = twentyfive[0].find('td',class_='time').get_text()
    listing['30K'] = thirty[0].find('td',class_='time').get_text()
    listing['35k'] = thirtyfive[0].find('td',class_='time').get_text()
    listing['40k'] = fourty[0].find('td',class_='time').get_text()
    listing['Finish'] = finish[0].find('td',class_='time').get_text()
    
    directory.append(listing)

    
    
    
    

    

    
```

```python
#creating dataframe with dictoranry
df = pd.DataFrame(directory)
df
```
